```python
import pandas as pd
import numpy as np
sp_df=pd.read_csv('data/israeli_supermarkets.csv')
sp_df.info()
```

```python
import numpy as np
import pandas as pd

# Drop rows without name or code
sp_df = sp_df.dropna(subset=['translated_itemname', 'itemcode'])

# Add review column with random int 1-5
sp_df['review'] = np.random.uniform(0.0, 5.0, size=len(sp_df)).round(1)

# Fill missing prices with random float (5.9–21.9)
sp_df['itemprice'] = sp_df['itemprice'].apply(
    lambda x: np.random.uniform(sp_df['itemprice'].min(),sp_df['itemprice'].max()) if pd.isna(x) else x
)

# Fillquantities with random int (0–15)
sp_df['itemquantity'] =np.random.randint(0, 15,size=len(sp_df))

sp_df.head(10)
```

```python
import numpy as np
import pandas as pd

# Assuming 'supermarket' is the column indicating supermarket name

def assign_promos_by_super(sp_df, promo_rate=0.2, promo_options=None):
    if promo_options is None:
        promo_options = [None, "10% off", "Buy 1 Get 1", "5 NIS discount", "5% off"]

    promo_flags = []
    promo_descs = []

    # Apply per supermarket
    for market, group in sp_df.groupby('supermarket'):
        n = len(group)
        # Number of promos to assign
        promo_n = int(np.round(n * promo_rate))
        promo_idx = np.random.choice(n, promo_n, replace=False)

        flag = np.zeros(n, dtype=int)
        flag[promo_idx] = 1

        # Assign descriptions only to promos, None otherwise
        desc = [None] * n
        for idx in promo_idx:
            desc[idx] = np.random.choice([p for p in promo_options if p])

        promo_flags.extend(flag)
        promo_descs.extend(desc)

    sp_df = sp_df.copy()
    sp_df['promo_flag'] = promo_flags
    sp_df['promo_desc'] = promo_descs
    return sp_df


sp_df = assign_promos_by_super(sp_df, promo_rate=0.2)
sp_df.head(10)
```

```python
#delete non relevant columns
sp_df=sp_df.drop(columns=['itemcode','itemname','manufacturername','manufacturecountry','manufactureitemdescription','unitqty'])
```
